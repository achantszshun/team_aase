```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import open3d as o3d

```

The plot below shows what our model is attemping to reproduce based on 2D slices visualized via the .raw files.

```python
def plot_point_cloud_from_ply(file_path, points_per_cloud=10000):
    """
    set your file path and then use as function input
    """
    # Load the .ply file as a mesh
    mesh = o3d.io.read_triangle_mesh(file_path)

    # Sample points from the mesh to create a point cloud
    point_cloud = mesh.sample_points_uniformly(number_of_points=points_per_cloud)
    points = np.asarray(point_cloud.points)  # Convert point cloud to numpy array

    # Plot the point cloud using Matplotlib
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=0.5)
    ax.set_title(f"Point Cloud from {file_path}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.show()
```

```python
file_path = "/content/scan_002.ply"
plot_point_cloud_from_ply(file_path)
```
