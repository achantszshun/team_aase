```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import open3d as o3d

```

```python
file_path = "/content/scan_001.ply"
mesh = o3d.io.read_triangle_mesh(file_path)

# Check if the mesh loaded correctly
if mesh.is_empty():
    print("Failed to load the mesh. Please check the file format and path.")
else:
    # Sample points from the mesh to create a point cloud
    point_cloud = mesh.sample_points_uniformly(number_of_points=10000)  # Adjust number of points for memory limits

    # Convert point cloud to numpy for visualization
    points = np.asarray(point_cloud.points)

    # Plot the point cloud using Matplotlib
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection="3d")
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=0.5)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.show()
```
